```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
# Load Beneficiary Data
train_beneficiary = pd.read_excel('/content/drive/My Drive/Train_Beneficiarydata-1542865627584.xlsx')
```

```python
# Find the midpoint of the dataset
midpoint = len(train_beneficiary) // 2

# Split into two halves
train = train_beneficiary.iloc[:midpoint]
test = train_beneficiary.iloc[midpoint:]

# Display the two halves
display("Training data:\n", train)
display("Testing Data:\n", test)
```

```python
# Replace null values in the 'DOD' column with "Alive"
train['DOD'].fillna("Alive", inplace=True)

# Display the updated DataFrame
display(train)
```

```python
# Display rows where 'DOD' does not contain "Alive"
non_alive_rows = train[train['DOD'] != "Alive"]

# Show the resulting DataFrame
display(non_alive_rows)
```

```python
# Check for duplicates in the 'BeneID' column of the train_beneficiary dataset
duplicates = train[train.duplicated(subset='BeneID')]

# Print the result
if not duplicates.empty:
    print(f"Found {len(duplicates)} duplicates in the 'BeneID' column of train_beneficiary.")
    print(duplicates)
else:
    print("No duplicates found in the 'BeneID' column of train_beneficiary.")

# Count the number of unique values in the 'BeneID' column of the train_beneficiary dataset
unique_beneid_count = train['BeneID'].nunique()

# Print the result
print(f"There are {unique_beneid_count} unique values in the 'BeneID' column of train_beneficiary.")
```

```python
# Rename the 'county' column to 'country'
train = train.rename(columns={'County': 'Country'})

# Verify the change
print(train.columns)
```

```python
# List of columns that need to be converted
chronic_conditions = [
    'ChronicCond_Alzheimer', 'ChronicCond_Heartfailure',
    'ChronicCond_KidneyDisease', 'ChronicCond_Cancer',
    'ChronicCond_ObstrPulmonary', 'ChronicCond_Depression',
    'ChronicCond_Diabetes', 'ChronicCond_IschemicHeart',
    'ChronicCond_Osteoporasis', 'ChronicCond_rheumatoidarthritis',
    'ChronicCond_stroke'
]

# Convert 2 to 0 and leave 1 as 1 for the specified columns
train[chronic_conditions] = train[chronic_conditions].replace({2: 0})

# Verify the changes
display(train[chronic_conditions].head())
```

```python
# Replace 0 with 'N' in the RenalDiseaseIndicator column
train['RenalDiseaseIndicator'] = train['RenalDiseaseIndicator'].replace({0: 'N'})

#Verify the changes
print(train_beneficiary['RenalDiseaseIndicator'].value_counts())
```

```python

# Replace 1 with 'M' and 2 with 'F' in the gender column for train_beneficiary
train_beneficiary['Gender'] = train_beneficiary['Gender'].replace({1: 'M', 2: 'F'})

# Verify the changes
print(train_beneficiary['Gender'].value_counts())
```
